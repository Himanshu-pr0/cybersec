```python
# Cell 1 — Install (optional) and imports
# Uncomment the next line only if you need to install packages in the notebook environment.
# !pip install pandas numpy scikit-learn matplotlib seaborn joblib

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# Plot style
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Cell 2 — Load data
DATA_PATH = "cybersecurity_attacks.csv"  # change path if needed

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATA_PATH} — place cybersecurity_attacks.csv in the notebook folder")

df = pd.read_csv(DATA_PATH)
print("Dataset shape:", df.shape)
display(df.head())
```

```python
# Cell 3 — Basic EDA: identify target column
# Adjust this if your label column has another name. We'll try common names.
target_col = 'Attack Type'

print("Using target column:", target_col)
print(df[target_col].value_counts().head(20))

# Plot distribution
plt.figure(figsize=(10,5))
sns.countplot(y=target_col, data=df, order=df[target_col].value_counts().index)
plt.title("Distribution of target classes")
plt.tight_layout()
plt.show()
```

```python
# Cell 4 — Preprocessing

# 1. Separate features and target
df_clean = df.copy()

# If target contains strings like 'normal' and named attacks, we will encode them using LabelEncoder later.
y_raw = df_clean[target_col]
X_raw = df_clean.drop(columns=[target_col])

# Drop high-cardinality columns
high_cardinality_cols = ['Timestamp', 'Source IP Address', 'Destination IP Address', 'Payload Data', 'User Information', 'Device Information', 'Geo-location Data', 'Proxy Information']
X_raw = X_raw.drop(columns=high_cardinality_cols)

# 2. Identify categorical vs numeric
cat_cols = X_raw.select_dtypes(include=['object', 'category']).columns.tolist()
num_cols = X_raw.select_dtypes(include=[np.number]).columns.tolist()

print("Categorical columns:", cat_cols)
print("Numerical columns:", num_cols)

# 3. Encode target using LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y_raw)
print("Target classes (LabelEncoder):", list(le.classes_))
print("Encoded target example distribution:", np.bincount(y))

# 4. Train-test split (stratify by label)
X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X_raw, y, test_size=0.2, random_state=42, stratify=y
)
print("Train shape:", X_train_raw.shape, "Test shape:", X_test_raw.shape)

# 5. Scale numeric features (fit scaler on training numeric cols)
scaler = StandardScaler()
if len(num_cols) > 0:
    X_train_num = scaler.fit_transform(X_train_raw[num_cols])
    X_test_num = scaler.transform(X_test_raw[num_cols])
    # replace numeric columns in copies
    X_train_scaled = X_train_raw.copy()
    X_test_scaled = X_test_raw.copy()
    X_train_scaled[num_cols] = X_train_num
    X_test_scaled[num_cols] = X_test_num
else:
    # no numeric columns — just copy
    X_train_scaled = X_train_raw.copy()
    X_test_scaled = X_test_raw.copy()

# 6. One-hot encode categorical features using pandas get_dummies
if len(cat_cols) > 0:
    X_train_ohe = pd.get_dummies(X_train_scaled, columns=cat_cols)
    X_test_ohe = pd.get_dummies(X_test_scaled, columns=cat_cols)
else:
    X_train_ohe = X_train_scaled.copy()
    X_test_ohe = X_test_scaled.copy()

# 7. Align test columns with train (add missing cols in test with 0, drop extra)
train_cols = X_train_ohe.columns
X_test_ohe = X_test_ohe.reindex(columns=train_cols, fill_value=0)

print("Final feature count:", len(train_cols))
print("Sample feature columns:", train_cols[:20])
```

```python
# Cell 5 — Train RandomForest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train_ohe, y_train)
print("RandomForest trained.")
```

```python
# Cell 6 — Evaluate
y_pred = rf.predict(X_test_ohe)
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {acc:.4f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=list(le.classes_)))

# Confusion matrix (plot)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=list(le.classes_), yticklabels=list(le.classes_))
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
# Cell 7 — Save artifacts
ARTIFACT_DIR = "artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

model_path = os.path.join(ARTIFACT_DIR, "rf_model.pkl")
scaler_path = os.path.join(ARTIFACT_DIR, "scaler.pkl")
labelenc_path = os.path.join(ARTIFACT_DIR, "label_encoder.pkl")
columns_path = os.path.join(ARTIFACT_DIR, "model_columns.pkl")

joblib.dump(rf, model_path)
joblib.dump(scaler, scaler_path)
joblib.dump(le, labelenc_path)
joblib.dump(list(train_cols), columns_path)

print("Saved model and preprocessors to", ARTIFACT_DIR)
print("- model:", model_path)
print("- scaler:", scaler_path)
print("- label encoder:", labelenc_path)
print("- model columns:", columns_path)
```

```python
# Cell 8 — Load artifacts and predict
# Load saved artifacts
rf_loaded = joblib.load(model_path)
scaler_loaded = joblib.load(scaler_path)
le_loaded = joblib.load(labelenc_path)
model_columns = joblib.load(columns_path)

# Define a helper to preprocess a raw dataframe to model-ready features
def preprocess_for_predict(df_input, numeric_cols=num_cols, categorical_cols=cat_cols, scaler_obj=scaler_loaded, model_cols=model_columns):
    """Return a numpy array aligned to model_cols (single row or multiple rows)"""
    df_proc = df_input.copy()
    # Scale numeric columns if present
    if len(numeric_cols) > 0:
        missing_num = [c for c in numeric_cols if c not in df_proc.columns]
        if missing_num:
            # fill missing numeric cols with 0
            for c in missing_num:
                df_proc[c] = 0
        df_proc[numeric_cols] = scaler_obj.transform(df_proc[numeric_cols])
    # One-hot encode categoricals
    if len(categorical_cols) > 0:
        df_proc = pd.get_dummies(df_proc, columns=[c for c in categorical_cols if c in df_proc.columns])
    # Align to model columns
    df_proc = df_proc.reindex(columns=model_cols, fill_value=0)
    return df_proc.values

# Use some rows from X_test_raw (original unprocessed dataframe) to demonstrate
# Recreate unprocessed test rows: we have X_test_raw (pandas DataFrame) from earlier
sample_raw = X_test_raw.iloc[:5].copy()  # first 5 test rows (raw, before scaling/ohe)
display(sample_raw.head())

X_sample = preprocess_for_predict(sample_raw)
preds = rf_loaded.predict(X_sample)
pred_probs = rf_loaded.predict_proba(X_sample)[:, 1] if hasattr(rf_loaded, "predict_proba") else None

# Decode human-readable labels
pred_labels = le_loaded.inverse_transform(preds)

print("Predicted labels:", pred_labels)
if pred_probs is not None:
    print("Predicted probabilities (attack class probability if label encoding maps attacks to class index 1):", pred_probs)
```

```python
# Cell 9 — Save sample predictions to CSV
out_df = sample_raw.copy().reset_index(drop=True)
out_df["pred_encoded"] = preds
out_df["pred_label"] = pred_labels
if pred_probs is not None:
    out_df["pred_prob_attack"] = pred_probs
out_df.to_csv(os.path.join(ARTIFACT_DIR, "sample_predictions.csv"), index=False)
print("Sample predictions saved to", os.path.join(ARTIFACT_DIR, "sample_predictions.csv"))
display(out_df.head())
```
